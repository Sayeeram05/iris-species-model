# Train And Test Data
    Data - X
    labels - y

```python
from sklearn.model_selection import train_test_split
```

```python
Input = Dataset['data']
Output = Dataset['target']
```

```python
X_train,X_test,y_train,y_test = train_test_split(Input,Output,random_state=0) 
```

# Random_state=0
    train and test data "will not" change if everytime i run this code
# test_size or train_size
    train and test data "will" change if everytime i run this code

# Bulid Model

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
knn = KNeighborsClassifier(n_neighbors=1)  # no of neighbors
```

# Train model

```python
knn.fit(X_train,y_train)
```

```python
#KNeighborsClassifier(algorithm="auto", leaf_size=30, metric="minkowski", metric_params=None, n_jobs=1, p=2, weights="uniform")
```

# Making Prediction

```python
new_data = np.array([[5, 2.9, 1, 0.2]])
```

```python
prediction = knn.predict(new_data)
```

```python
prediction
```

# Evaluating The Model

```python
predict_test = knn.predict(X_test)
predict_test
```

```python
np.mean(predict_test == y_test)
```
